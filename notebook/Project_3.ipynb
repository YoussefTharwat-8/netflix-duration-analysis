```python
#lib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read_data
data = pd.read_csv('../data/netflix_titles.csv')
data.head()
```

```python
#data info
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# filter movies
data = data[data['type'] == 'Movie']
```

```python
data.head()
```

```python
data = data.dropna(subset=['duration'])
data['duration'] = data['duration'].str.replace(' min', '')
data['duration'] = data['duration'].astype(int)
```

```python
data.head()
```

```python
data= data[data['release_year'] >= 2000]
data['release_year'].value_counts().sort_index()
```

```python
data['duration'].describe()
```

```python
#remove outliers
data = data[(data['duration'] > 40) & (data['duration'] < 210)]
```

```python
data['duration'].describe()
```

```python
data[['release_year', 'duration']].isnull().sum()
```

```python
data
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12,8))
sns.regplot(data=data,x='release_year',y='duration',x_jitter=0.4,scatter_kws={'alpha':0.3,'color':'#E50914','s':15},line_kws={'color':'black','linewidth':2})
plt.title('Release Year vs Duration')
plt.xlabel('Release Year')
plt.ylabel('Duration (min)')
plt.show()
```

```python
data['genre']=data['listed_in'].str.split(',').str[0].str.strip()
top_10_genres=data['genre'].value_counts().head(10).index
top_genre_data=data[data['genre'].isin(top_10_genres)]
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12,8))
sns.boxenplot(data=top_genre_data,y='duration',x='genre',palette='muted',order=top_genre_data.groupby('genre')['duration'].median().sort_values().index)
plt.title('Movies Duration by Top 10 genres')
plt.xlabel('Genre')
plt.ylabel("Duration (min)")
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
data_with_directors = data.dropna(subset=['director'])
top_10_directors = data_with_directors['director'].value_counts().head(10).index
top_dir_data = data_with_directors[data_with_directors['director'].isin(top_10_directors)]
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12, 8))
sns.boxplot(data=top_dir_data,x='duration',y='director',palette='Set2',order=top_dir_data.groupby('director')['duration'].median().sort_values().index)
plt.title('Movie Duartion by Director')
plt.xlabel('Duration (min)')
plt.ylabel('Director')
plt.tight_layout()
plt.show()
```
